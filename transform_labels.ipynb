```python
import os
import xml.etree.ElementTree as ET

input_dir = r"C:\Users\João Matias\Desktop\UA_DETRAC_2016\TrainSet"
output_dir = r"C:\Users\João Matias\Desktop\test-labels (1)\test-labels"

image_width = 1920
image_height = 1080
types = {"car": 0,
         "bus": 1,
         "van": 3,
         "others": 4}

for filename in os.listdir(input_dir):
    if filename.endswith("xml"):
        input_file = os.path.join(input_dir, filename)
        video_name = os.path.splitext(os.path.basename(input_file))[0]

    tree = ET.parse(input_file)
    root = tree.getroot()

    for frame in root.iter("frame"):
        frame_id = int(frame.attrib["num"])
        target_list = frame.find("target_list")

        annotations = []

        for target in target_list.findall("target"):
            attribute = target.find("attribute")
            box = target.find("box")

            left = float(box.attrib["left"]) / 2
            top = float(box.attrib["top"]) / 2
            width = float(box.attrib["width"]) / 2
            height = float(box.attrib["height"]) / 2
            class_name = attribute.attrib["vehicle_type"]
            
            print(f"id: {frame_id}, class: {class_name}, left: {left}, top: {top}, width: {width}, height: {height}")

            x_norm = (left + width / 2) / 960
            y_norm = (top + height / 2) / 540
            width_norm = width / 960
            height_norm = height / 540
            class_id = types.get(class_name, 4)

            print(f"class: {class_id}, x: {x_norm}, y: {y_norm}, width_norm: {width_norm}, height_norm: {height_norm}")
            annotations.append(f"{class_id} {x_norm:0.6f} {y_norm:.6f} {width_norm:.6f} {height_norm:.6f}")
        
        filename = f"{video_name}_frame{frame_id:05d}.txt"
        filepath = os.path.join(output_dir, filename)

        with open(filepath, "w") as f:
            for annotation in annotations:
                f.write(annotation + "\n")




```
